## Задание №1

##### Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
import os
os.chdir('C:/Users/USER 2/Documents/lit') 
with open('literary-anecdotes.txt', 'r', encoding='utf-8') as f: 
    text = f.read()

from string import punctuation
tex = [item.strip(punctuation) for item in text.split()]
te = ' '.join(tex)

from pymystem3 import Mystem 
m = Mystem() 
lemmas = m.lemmatize(te) 
anec = ''.join(lemmas)
print(anec)

import json 
with open('anecdotes.json', 'w', encoding='utf-8') as g: 
    json.dump(anec, g, ensure_ascii=False) 
with open('anecdotes.json', 'r', encoding='utf-8') as h: 
    anec = h.read()
```

## Задание №2

##### Очистите лемматизированный текст от стоп-слов и посчитайте ipm для оставшихся. Выведите 20 самых частотных по этому параметру слов.

```python
os.chdir('C:/Users/USER 2/Documents/lit') 
with open('rus_stopwords.txt', 'r', encoding='utf-8') as i: 
    stop = i.read()
an = anec.split()
ipm = {}
for word in an: 
    if word in stop: 
        an.remove(word)
    else:
        ipm[word] = (an.count(word) / len(an)) * 1000000
sorted_ipm = sorted(ipm.items(), key=lambda x: x[1], reverse=True)
print(sorted_ipm[:20])
```
