```python
# №1
sen = "варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг"
# Разбиваем текст на слова
sp = sen.split()
# Создаем список биграмм (на каждой итерации цикла печатаем биграмму и добавляем ее в список)
bi = []
for i in range(len(sp)-1):
    a = sp[i] + ' ' + sp[i+1]
    bi.append(a)
print(bi)
```

```python
# №3

# 1 (сохранение текста в переменную)
poem = """Удивительная кошка

Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!"""

# 2 (длина текста в символах)
print(len(poem), '\n')

# 3 (очищаем текст от пунктуации с помощью list comprehensions)
from string import punctuation
print([item.strip(punctuation) for item in poem.split()],'\n')

# 4 (объединяем в строку)
te = [item.strip(punctuation) for item in poem.split()]
print(' '.join(te), '\n')

# 5 (проверяем, в каком регистре слово, и приводим к нижнему все)
te1 = [word if word.islower() else word.lower() for word in te]
print(te1, '\n')
    
# 6 (длина текста в словах)
print(len(te), '\n')

# 7 (кол-во уникальных слов)
poem1 = set(poem.split())
print(len(poem1),'\n')

# 8 (последние 10 слов)
print(te1[-10:], '\n')

# 9 (объединяем в строку полученный в п.3 список; слова через знак переноса строки)
print('\n'.join(te), '\n')
poem2 = '\n'.join(te)

# 10 (выводим символы с 20-го по 30-й)
tex = list(poem2)
print(tex[21:32], '\n')

# 11 (выводим биграммы и кол-во употреблений каждой)
bi1 = []
for i in range(len(te1)-1):
    a = te1[i] + ' ' + te1[i+1]
    bi1.append(a)
for a in bi1:
    print(a,'-', bi1.count(a), 'раз')
```
