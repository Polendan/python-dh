##### Пройтись циклом по рабочей директории, читая содержимое каждого файла и записывая его в новый файл poems.txt По завершении цикла должен получиться файл со всеми стихотворениями!
for root, dirs, files in os.walk('C:/Users/Danya/Documents/poems'):
    for name in files:
        fullname = os.path.join(root, name)
        f = open(fullname, 'r', encoding='utf-8')
        poem1 = f.read()
        with open('poems.txt', 'a', encoding='utf-8') as f:
            f.write(poem1 + '\n\n')
